```python
def mask2rle(img):
    '''
    img: numpy array, 1 -> mask, 0 -> background
    Returns run length as string formated
    '''
    pixels= img.T.flatten()
    pixels = np.concatenate([[0], pixels, [0]])
    runs = np.where(pixels[1:] != pixels[:-1])[0] + 1
    runs[1::2] -= runs[::2]
    return ' '.join(str(x) for x in runs)

def make_mask(row_id, df):
    '''Given a row index, return image_id and mask (256, 1600, 4) from the dataframe `df`'''
    fname = df.iloc[row_id].name
    labels = df.iloc[row_id][:4]
    masks = np.zeros((256, 1600, 4), dtype=np.float32) # float32 is V.Imp
    # 4:class 1～4 (ch:0～3)

    for idx, label in enumerate(labels.values):
        if label is not np.nan:
            label = label.split(" ")
            positions = map(int, label[0::2])
            length = map(int, label[1::2])
            mask = np.zeros(256 * 1600, dtype=np.uint8)
            for pos, le in zip(positions, length):
                mask[pos:(pos + le)] = 1
            masks[:, :, idx] = mask.reshape(256, 1600, order='F')
    return fname, masks
```

```python
# Dataloader

class SteelDataset(Dataset):
    def __init__(self, df, data_folder, mean, std, phase):
        self.df = df
        self.root = data_folder
        self.mean = mean
        self.std = std
        self.phase = phase
        self.transforms = get_transforms(phase, mean, std)
        self.fnames = self.df.index.tolist()

    def __getitem__(self, idx):
        image_id, mask = make_mask(idx, self.df)
        image_path = os.path.join(self.root, "train_images",  image_id)
        img = cv2.imread(image_path)
        augmented = self.transforms(image=img, mask=mask)
        img = augmented['image']
        mask = augmented['mask'] # 1x256x1600x4
        mask = mask[0].permute(2, 0, 1) # 4x256x1600
        return img, mask

    def __len__(self):
        return len(self.fnames)


def get_transforms(phase, mean, std):
    list_transforms = []
    if phase == "train":
        list_transforms.extend(
            [
                HorizontalFlip(p=0.5), # only horizontal flip as of now
            ]
        )
    list_transforms.extend(
        [
            Normalize(mean=mean, std=std, p=1),
            ToTensor(),
        ]
    )
    list_trfms = Compose(list_transforms)
    return list_trfms

def provider(
    data_folder,
    df_path,
    phase,
    mean=None,
    std=None,
    batch_size=8,
    num_workers=4,
):
    '''Returns dataloader for the model training'''
    df = pd.read_csv(df_path)
    # https://www.kaggle.com/amanooo/defect-detection-starter-u-net
    #df['ImageId'], df['ClassId'] = zip(*df[['ImageId','ClassId']])
    df['ClassId'] = df['ClassId'].astype(int)
    df = df.pivot(index='ImageId',columns='ClassId',values='EncodedPixels')
    df['defects'] = df.count(axis=1)
    
    train_df, val_df = train_test_split(df, test_size=0.2, stratify=df["defects"], random_state=69)
    df = train_df if phase == "train" else val_df
    image_dataset = SteelDataset(df, data_folder, mean, std, phase)
    dataloader = DataLoader(
        image_dataset,
        batch_size=batch_size,
        num_workers=num_workers,
        pin_memory=True,
        shuffle=True,   
    )

    return dataloader
```

```python
# Utility functions (Dice and IoU metric implementations, metric logger for training and validation)

def predict(X, threshold):
    '''X is sigmoid output of the model'''
    X_p = np.copy(X)
    preds = (X_p > threshold).astype('uint8')
    return preds

def metric(probability, truth, threshold=0.5, reduction='none'):
    '''Calculates dice of positive and negative images seperately'''
    '''probability and truth must be torch tensors'''
    batch_size = len(truth)
    with torch.no_grad():
        probability = probability.view(batch_size, -1)
        truth = truth.view(batch_size, -1)
        assert(probability.shape == truth.shape)

        p = (probability > threshold).float()
        t = (truth > 0.5).float()

        t_sum = t.sum(-1)
        p_sum = p.sum(-1)
        neg_index = torch.nonzero(t_sum == 0)
        pos_index = torch.nonzero(t_sum >= 1)

        dice_neg = (p_sum == 0).float()
        dice_pos = 2 * (p*t).sum(-1)/((p+t).sum(-1))

        dice_neg = dice_neg[neg_index]
        dice_pos = dice_pos[pos_index]
        dice = torch.cat([dice_pos, dice_neg])

#         dice_neg = np.nan_to_num(dice_neg.mean().item(), 0)
#         dice_pos = np.nan_to_num(dice_pos.mean().item(), 0)
#         dice = dice.mean().item()

        num_neg = len(neg_index)
        num_pos = len(pos_index)

    return dice, dice_neg, dice_pos, num_neg, num_pos

class Meter:
    '''A meter to keep track of iou and dice scores throughout an epoch'''
    def __init__(self, phase, epoch):
        self.base_threshold = 0.5 # <<<<<<<<<<< here's the threshold
        self.base_dice_scores = []
        self.dice_neg_scores = []
        self.dice_pos_scores = []
        self.iou_scores = []

    def update(self, targets, outputs):
        probs = torch.sigmoid(outputs)
        dice, dice_neg, dice_pos, _, _ = metric(probs, targets, self.base_threshold)
        self.base_dice_scores.extend(dice.tolist())
        self.dice_pos_scores.extend(dice_pos.tolist())
        self.dice_neg_scores.extend(dice_neg.tolist())
        preds = predict(probs, self.base_threshold)
        iou = compute_iou_batch(preds, targets, classes=[1])
        self.iou_scores.append(iou)

    def get_metrics(self):
        dice = np.nanmean(self.base_dice_scores)
        dice_neg = np.nanmean(self.dice_neg_scores)
        dice_pos = np.nanmean(self.dice_pos_scores)
        dices = [dice, dice_neg, dice_pos]
        iou = np.nanmean(self.iou_scores)
        return dices, iou

def epoch_log(phase, epoch, epoch_loss, meter, start):
    '''logging the metrics at the end of an epoch'''
    dices, iou = meter.get_metrics()
    dice, dice_neg, dice_pos = dices
    print("Loss: %0.4f | IoU: %0.4f | dice: %0.4f | dice_neg: %0.4f | dice_pos: %0.4f" % (epoch_loss, iou, dice, dice_neg, dice_pos))
    return dice, iou

def compute_ious(pred, label, classes, ignore_index=255, only_present=True):
    '''computes iou for one ground truth mask and predicted mask'''
    pred[label == ignore_index] = 0
    ious = []
    for c in classes:
        label_c = label == c
        if only_present and np.sum(label_c) == 0:
            ious.append(np.nan)
            continue
        pred_c = pred == c
        intersection = np.logical_and(pred_c, label_c).sum()
        union = np.logical_or(pred_c, label_c).sum()
        if union != 0:
            ious.append(intersection / union)
    return ious if ious else [1]

def compute_iou_batch(outputs, labels, classes=None):
    '''computes mean iou for a batch of ground truth masks and predicted masks'''
    ious = []
    preds = np.copy(outputs) # copy is imp
    labels = np.array(labels) # tensor to np
    for pred, label in zip(preds, labels):
        ious.append(np.nanmean(compute_ious(pred, label, classes)))
    iou = np.nanmean(ious)
    return iou
```

# 4. TRAINING AND VALIDATION

```python
def best_entropy(p,target,ths=0.5):
    
    # NegEntropy
    a = p * np.log(p + 1e-8)
    b = +(1-p) * np.log(1-p + 1e-8)
    mask_a = np.isfinite(a)
    mask_b = np.isfinite(b)
    filtered_arr_a = a[mask_a]
    filtered_arr_b = b[mask_b]
    NegEntropy = np.nanmean(filtered_arr_a) + np.nanmean(filtered_arr_b)
    
    # BCE
    a = target * np.log(p + 1e-8)
    b = (1-target) * np.log(1-p + 1e-8)
    mask_a = np.isfinite(a)
    mask_b = np.isfinite(b)
    filtered_arr_a = a[mask_a]
    filtered_arr_b = b[mask_b]    
    BCE = np.nanmean(filtered_arr_a) + np.nanmean(filtered_arr_b)
    
    return BCE - NegEntropy
```

```python
class Trainer(object):
    '''This class takes care of training and validation of our model'''
    def __init__(self, model, device='cuda'):
        self.num_workers = 6
        self.batch_size = {"train": 2, "val": 2}
        self.accumulation_steps = 32 // self.batch_size['train']
        self.lr = 5e-4
        self.num_epochs = 20
        self.best_iou = -float("inf")
        self.phases = ["train", "val"]
        self.device = device
        #torch.set_default_tensor_type("torch.cuda.FloatTensor")
        self.net = model
        self.net_copy = deepcopy(self.net)
        self.net = self.net.to(self.device)
        self.criterion = torch.nn.BCEWithLogitsLoss()
        #self.optimizer = optim.Adam(self.net.parameters(), lr=self.lr)
        self.optimizer = FastAdaBelief(self.net.parameters())
        self.scheduler = ReduceLROnPlateau(self.optimizer, mode="min", patience=3, verbose=True)
        
        cudnn.benchmark = True
        self.dataloaders = {
            phase: provider(
                data_folder=data_folder,
                df_path=train_df_path,
                phase=phase,
                mean=(0.485, 0.456, 0.406),
                std=(0.229, 0.224, 0.225),
                batch_size=self.batch_size[phase],
                num_workers=self.num_workers,
            )
            for phase in self.phases
        }
        self.losses = {phase: [] for phase in self.phases}
        self.iou_scores = {phase: [] for phase in self.phases}
        self.dice_scores = {phase: [] for phase in self.phases}
        self.early_stopping_warmup = 0
        self.early_stopping_phase_2 = 20
    
    def forward(self, images, targets):
        images = images.to(self.device)
        masks = targets.to(self.device)
        outputs = self.net(images)
        loss = self.criterion(outputs, masks)
        return loss, outputs

    def iterate(self, epoch, phase):
        meter = Meter(phase, epoch)
        start = time.strftime("%H:%M:%S")
        print(f"Starting epoch: {epoch} | phase: {phase} | ⏰: {start}")
        batch_size = self.batch_size[phase]
        self.net.train(phase == "train")
        self.criterion = torch.nn.BCEWithLogitsLoss()
        self.optimizer = optim.Adam(self.net.parameters(), lr=self.lr)
        self.scheduler = ReduceLROnPlateau(self.optimizer, mode="min", patience=3, verbose=True)
        
        dataloader = self.dataloaders[phase]
        running_loss = 0.0
        total_batches = len(dataloader)

        self.optimizer.zero_grad()
        for itr, batch in tqdm(enumerate(dataloader), total=len(dataloader)): # replace `dataloader` with `tk0` for tqdm            
            images, targets = batch
            loss, outputs = self.forward(images, targets)
            loss = loss / self.accumulation_steps
            if phase == "train":
                loss.backward()
                if (itr + 1 ) % self.accumulation_steps == 0:
                    self.optimizer.step()
                    self.optimizer.zero_grad()
            running_loss += loss.item()
            outputs = outputs.detach().cpu()
            meter.update(targets, outputs)

        epoch_loss = (running_loss * self.accumulation_steps) / total_batches
        dice, iou = epoch_log(phase, epoch, epoch_loss, meter, start)
        self.losses[phase].append(epoch_loss)
        self.dice_scores[phase].append(dice)
        self.iou_scores[phase].append(iou)
        torch.cuda.empty_cache()
        return epoch_loss, iou
    
    # modifciations here  
    def iterate_get_preds_and_labels(self, phase="val"):
        """obtenir les predicitons et les labels dans la meme output"""
        self.batch_size = {"train": 1, "val": 4}
        self.dataloaders = {
            phase: provider(
                data_folder=data_folder,
                df_path=train_df_path,
                phase=phase,
                mean=(0.485, 0.456, 0.406),
                std=(0.229, 0.224, 0.225),
                batch_size=self.batch_size[phase],
                num_workers=self.num_workers,
            )
            for phase in self.phases
        }
        outputs_pred_label = []
        batch_size = self.batch_size[phase]
        self.net.train(phase == "train")
        dataloader = self.dataloaders[phase]
        total_batches = len(dataloader)
        self.optimizer.zero_grad()
        liste_dir = sorted(os.listdir('data_predictions/'))
        for itr, name_file in tqdm(enumerate(liste_dir), total=len(liste_dir)):
             images,outputs,targets  = load_compressed_pickle('data_predictions/' + name_file)
            images,outputs,targets = torch.Tensor(images),torch.Tensor(outputs),torch.Tensor(targets)
            _, outputs = self.forward(images, targets)
            images = images.detach().cpu().numpy()
            outputs = torch.softmax(outputs, dim=1).detach().cpu().numpy()
            targets = targets.detach().cpu().numpy()
            save_compressed_pickle([images,outputs,targets],'data_predictions/'+name_file) 

    def get_topk_errors_corrected(self, topk=5):
        liste_dir = sorted(os.listdir('data_predictions/'))
        liste_entropy = []
        liste_dir_ = []
        for itr, name_file in tqdm(enumerate(liste_dir), total=len(liste_dir)):
            images,outputs,targets = load_compressed_pickle('data_predictions/' + name_file)
            for i in range(len(images)):
                output, target = outputs[i],targets[i]
                for j in range(4):
                    entropy = best_entropy(output[j],target[j],ths=0.5)
                    liste_entropy+= [entropy]
                    liste_dir_.append('data_predictions/'+name_file)
        indexes_sorted = np.argsort(liste_entropy)[-topk:]

        for i in tqdm(indexes_sorted):
            print(i)
            file_name = liste_dir_[i]
            index_image = 0
            index_target = i%4
            images,outputs,targets = load_compressed_pickle('data_predictions/'+file_name)
            
            targets[index_image,index_target] = outputs[index_image,index_target]
            save_compressed_pickle([images,outputs,targets],'data_predictions/'+file_name) 
            print(str(i)+' corrected')
          
    def retrain(self, epoch, phase):
        meter = Meter(phase, epoch)
        start = time.strftime("%H:%M:%S")
        print(f"Starting epoch: {epoch} | phase: {phase} | ⏰: {start}")
        batch_size = self.batch_size[phase]
        self.net.train(phase == "train")
        running_loss = 0.0
        self.criterion = torch.nn.BCEWithLogitsLoss()
        self.optimizer = optim.Adam(self.net.parameters(), lr=self.lr)
        self.scheduler = ReduceLROnPlateau(self.optimizer, mode="min", patience=3, verbose=True)
        self.optimizer.zero_grad()
        liste_dir = os.listdir('data_predictions/')
        liste_dir = random.sample(liste_dir, len(liste_dir))
        for itr, name_file in tqdm(enumerate(liste_dir), total=len(liste_dir)):
            images, outputs, targets = load_compressed_pickle('data_predictions/'+name_file)
            images, outputs, targets = torch.Tensor(images), torch.Tensor(outputs), torch.Tensor(targets)
            loss, outputs = self.forward(images, targets)
            loss = loss / self.accumulation_steps
            if phase == "train":
                loss.backward()
                if (itr + 1 ) % self.accumulation_steps == 0:
                    self.optimizer.step()
                    self.optimizer.zero_grad()
            running_loss += loss.item()
            outputs = outputs.detach().cpu()
            meter.update(targets, outputs)
        
        epoch_loss = (running_loss * self.accumulation_steps) / len(liste_dir)
        dice, iou = epoch_log(phase, epoch, epoch_loss, meter, start)
        self.losses[phase].append(epoch_loss)
        self.dice_scores[phase].append(dice)
        self.iou_scores[phase].append(iou)
        torch.cuda.empty_cache()
        return epoch_loss

    def predict_data(self,phase='train'):

        dataloader = self.dataloaders[phase]
        for itr, batch in tqdm(enumerate(dataloader), total=len(dataloader)):
            images, targets = batch
            
            with torch.no_grad():
                _, outputs = self.forward(images, targets)

            images = images.detach().cpu().numpy()
            outputs = outputs.detach().cpu().numpy()
            targets = targets.detach().cpu().numpy()
            save_compressed_pickle([images,outputs,targets],'data_predictions/batch_data_predicitons_{}.pkl'.format(itr))
        
    def start(self):
        ok = 0
        while True:
            if os.path.isfile('data_predictions2/batch_data_predicitons_0.pkl'):
                start = 1
                copytree('data_predictions2/', 'data_predictions/')
            else:
                start = 0

            self.net = deepcopy(self.net_copy)
            self.net = self.net.to(self.device)
            
            # PHASE I warmup https://github.com/D641593/MixNet
            early_stopping = 0
            epoch = 0
            while early_stopping < self.early_stopping_warmup:
                if start == 0:
                    self.iterate(epoch, "train")
                else:
                    self.retrain(epoch, "train")
                
                state = {
                    "epoch": epoch,
                    "best_iou": self.best_iou,
                    "state_dict": self.net.state_dict(),
                    "optimizer": self.optimizer.state_dict(),
                }
                with torch.no_grad():
                    val_loss, iou = self.iterate(epoch, "val")
                    self.scheduler.step(val_loss)
                if iou > self.best_iou:
                    print("******** New optimal found, saving state ********")
                    state["loss"] = val_loss
                    state["best_iou"] = self.best_iou
                    torch.save(state, "./model.pth")
                    early_stopping = 0
                    self.best_iou = iou
                else:
                    early_stopping += 1 
                epoch += 1
            
            if epoch == 0 and start == 0:
                var = torch.load('model.pth')
                self.best_iou = var['best_iou']
                self.net.load_state_dict(var['state_dict'])
                self.predict_data(phase='train')
                print('initial performances :', self.best_iou)

            # PHASE II Active learning
            print('PHASE 2')
            early_stopping = 0
            while early_stopping < self.early_stopping_phase_2:
                # PHASE IIA predictions on train dataset 
                print('PHASE IIA')
                self.iterate_get_preds_and_labels(phase="train")
                # PHASE IIB get correction on topk confidence entropy training elements
                print('PHASE IIB')
                self.get_topk_errors_corrected(topk=5)
                print('PHASE IIC')
                # PHASE IIC training epoch on corrected data
                self.retrain(epoch, phase='train')

                state = {
                    "epoch": epoch,
                    "best_iou": self.best_iou,
                    "state_dict": self.net.state_dict(),
                    "optimizer": self.optimizer.state_dict(),
                }

                with torch.no_grad():
                    val_loss, iou = self.iterate(epoch, "val")
                    self.scheduler.step(val_loss)
                if iou > self.best_iou:
                    print("******** New optimal found, saving state ********")
                    state["loss"] = val_loss
                    state['best_iou'] = self.best_iou
                    torch.save(state, "./model_.pth")
                    copytree('data_predictions/', 'data_predictions2/')
                    early_stopping = 0
                    self.best_iou = iou
                else:
                    early_stopping += 1 
                early_stopping += 1
                epoch += 1 
            ok = 1
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
def dice_coeff(outputs, targets, threshold=0.5, smooth=1e-6):
    """
    Calcule le coefficient de Dice entre les sorties et les cibles.

    Args:
    outputs (np.ndarray): tableau contenant les prédictions du modèle.
    targets (np.ndarray): tableau contenant les étiquettes vraies.

    Returns:
    float: le coefficient de Dice.
    """
    
    # Aplatir les tableaux
    outputs_flat = (outputs.reshape(-1) > threshold).astype(np.int)
    targets_flat = (targets.reshape(-1) > threshold).astype(np.int)
    
    # Calculer l'intersection
    intersection = (outputs_flat * targets_flat).sum()

    # Calculer le coefficient de Dice
    dice = (2. * intersection + smooth) / (outputs_flat.sum() + targets_flat.sum() + smooth)

    return dice
```

```python
def iou_coeff(outputs, targets, threshold=0.5, smooth=1e-6):
    """
    Calculer l'intersection sur l'union (IoU) de deux masques.
    
    :param output: masque de sortie (numpy array)
    :param target: masque cible (numpy array)
    :return: la valeur IoU
    """
    # Calculer l'intersection
    outputs_flat = (outputs.reshape(-1) > threshold).astype(np.int)
    targets_flat = (targets.reshape(-1) > threshold).astype(np.int)
    intersection = np.logical_and(outputs_flat, targets_flat)
    
    # Calculer l'union
    union = np.logical_or(outputs_flat, targets_flat)
    
    # Calculer l'IoU
    iou = np.sum(intersection) / np.sum(union)
    
    return iou
```

```python
class Meter:
    '''A meter to keep track of iou and dice scores throughout an epoch'''
    def __init__(self, threshold=0.5):
        self.base_threshold = threshold
        self.base_dice_scores = []
        self.dice_neg_scores = []
        self.dice_pos_scores = []
        self.iou_scores = []

    def update(self, targets, outputs):
        probs = sigmoid(outputs)
        dice = dice_coeff(probs, targets, self.base_threshold)
        self.base_dice_scores.append(dice)

    def get_metrics(self):
        dice = np.nanmean(self.base_dice_scores)
        return dice
```

```python
def metric_with_fix_ths(folder='data_predictions/', func=None):
    liste_elements = os.listdir(folder)
    metric = Meter(threshold=0.5)
    for itr,batch_file in tqdm(enumerate(liste_elements),total=len(liste_elements)):
        with open(folder + batch_file, 'rb') as f:
            _, outputs, targets = pickle.load(f)
        metric.update(outputs,targets)
    return metric
```

```python
def get_mask_prob(outputs, threshold=0.95):
    """
    fonction qui créé les mask dilaté de 1 contenant les probas de chaque instance de l'output
    """
    liste_masks = []
    outputs = sigmoid(outputs)
    for outputs_ in outputs:
        liste_mask = []
        for output in outputs_:
            output_ = (output>threshold).astype(np.uint8)
            contours, hierarchy = cv2.findContours(output_, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
            liste_elements = list(contours)
            output_image = cv2.cvtColor(output_ * 255, cv2.COLOR_GRAY2BGR)
            for contour in liste_elements:
                mask = np.zeros(output.shape, dtype=np.uint8)
                cv2.drawContours(mask, [contour], -1, (1), thickness=cv2.FILLED)
                kernel = np.ones((3,3), np.uint8) 
                mask_dilated = cv2.dilate(mask, kernel, iterations=1)
                liste_mask.append(output*mask)
        liste_masks.append(liste_mask)
    return liste_masks 
```

```python
def get_mask_target(targets, threshold=0.5):
    """
    fonction qui créé les mask dilaté de 1 contenant les probas de chaque instance de l'output
    """
    liste_masks = []
    for targets_ in targets:
        liste_mask = []
        for target in targets_:
            target_ = (target>threshold).astype(np.uint8)
            contours, hierarchy = cv2.findContours(target_, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
            liste_elements = list(contours)
            output_image = cv2.cvtColor(target_ * 255, cv2.COLOR_GRAY2BGR)  # Convert to BGR image for color drawing
            for contour in liste_elements:
                mask = np.zeros(target.shape, dtype=np.uint8)
                cv2.drawContours(mask, [contour], -1, (1), thickness=cv2.FILLED)
                kernel = np.ones((3,3), np.uint8)
                mask_dilated = cv2.dilate(mask, kernel, iterations=1)
                liste_mask.append(target*mask)
        liste_masks.append(liste_mask)
    return liste_masks 
```

```python
def get_best_couples(outputs_masks, targets_masks):
    maximum = 0
    for i in range(len(outputs_masks)):
        output = outputs_masks[i]
        for j in range(len(targets_masks)):
            target = targets_masks[j]
            dice = dice_coeff(output,target)
            if dice> maximum:
                maximum=dice
                indexes = (i,j)
    return indexes

def best_couple_from_dice(outputs_masks, targets_masks):
    outputs, targets = [], []
    while len(outputs_masks)!=0 and len(targets_masks)!=0:
        i,j = get_best_couples(outputs_masks, targets_masks)
        outputs.append(outputs_masks.pop(i))
        targets.append(targets_masks.pop(j))
    return outputs,targets
```

```python
outputs_final, targets_final, images_final = [],[],[]
liste_files = os.listdir('data_predictions/')
for itr,file in tqdm(enumerate(liste_files), total=len(liste_files)):
    with open('data_predictions/'+file, 'rb') as f:
        images, outputs, targets = pickle.load(f)
    targets_masks = get_mask_target(targets)
    outputs_masks = get_mask_prob(outputs, threshold=0.5)
    for i in range(len(targets_masks)):
        outputs, targets = outputs_masks[i], targets_masks[i]
        outputs_, targets_ = best_couple_from_dice(outputs, targets)

        outputs_final += outputs_
        targets_final += targets_
        images_final += list(images[i])
```

```python
liste_files_names = os.listdir('data_predictions/')

def compute_dice(output_mask, target_mask):
    dice_total = []
    for mask_t, mask_o in tqdm(zip(target_mask, output_mask)):
        dice_total.append(dice_coeff(mask_o,mask_t))
    return dice_total

def compute_iou(output_mask, target_mask):
    dice_total = []
    for mask_t, mask_o in tqdm(zip(target_mask, output_mask)):
        dice_total.append(iou_coeff(mask_o,mask_t))
    return dice_total

def compute_entropy(output_mask, target_mask, func=None):
    entropy_total = []
    for mask_t, mask_o in tqdm(zip(target_mask, output_mask)):
        entropy_total.append(func(mask_o,0.5))
    return entropy_total
```
